## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
```

```python
logreg.intercept_
```

```python
logreg.coef_[0]
```

```python
# df_coef = pd.DataFrame{}
```

```python
df_coef = pd.DataFrame({'feature_names':X_train.columns, 
                        'coef':logreg.coef_[0]})
df_coef
```

```python
df_coef.index = df_coef.index+1
df_coef.loc[0] = ['intercept', logreg.intercept_[0]]
df_coef = df_coef.sort_index()
df_coef
```

## Model Evaluation 

```python
pred = logreg.predict(X_test)
proba = logreg.predict_proba(X_test)
```

```python
len(pred)
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(plot_confusion_matrix(logreg, X_test,y_test, values_format='d'))
```

- 0 = Bad Borrower
- 1 = Good Borrower


- TN = Actually bad borrower, predicted as bad borrower (TRUE) = 6
- FP = Actually bad borrower, predicted as good borrower (FALSE) = 10184 
- FN = Actually good borrower, predicted as bad borrower (FALSE) = 4
- TP =  Actually good borrower, predicted as good borrower (TRUE) = 83063



- In this case, our model has missclasify 10184 + 4 people/borrower.FP = 10184, FN = 6. 
- We'll Focus on FP
- Having FP higher than FN is something we dont want to happen, because it means we will give a credit to the borrower will be default in the feature. If that is happen the bank or creditor will suffer high looss.
- For Metrics we'll focus on Precision(+) & Recall(-) instead of stick with overall accuracy.

```python
print(classification_report(y_test,pred))
```

```python
mx = confusion_matrix(y_test,pred)
mx
```

```python
np.reshape(mx,(4,1)).tolist()
```

```python
df_class = pd.DataFrame({'Class':['TN','FP','FN','TP'], 
                         'values':[i[0] for i in np.reshape(mx,(4,1)).tolist()]})
df_class
```

```python
Prec = 83063 / (83063+10184)
Rec = 6 / (6+10184)

print(f'Precision(+) : {Prec}')
print(f'Recall(-) : {Rec}')
```

## Try Threshold

```python
df_pred = pd.DataFrame({'actual':y_test['good_bad'], 
             'prediction':pred, 
             'proba':proba[:,1]})
df_pred
```

```python
df_thresh = df_pred[(df_pred['actual']==0)&(df_pred['prediction']==1)]
df_thresh
```

```python
# try adjust proba >0.90
pred_over_90 = []
for p in proba[:,1]:
    if p >= 0.90:
        pred_over_90.append(1)
    else:
        pred_over_90.append(0)
```

```python
print(confusion_matrix(y_test,pred_over_90))
```

```python
print(classification_report(y_test,pred_over_90))
```
