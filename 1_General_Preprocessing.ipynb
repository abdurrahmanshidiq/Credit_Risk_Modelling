```python
# ALTERNATIVE WAY

def fix_date(x):
    if pd.notnull(x):
        x = str(x)
        year = x[:4]
        month = x[5:7]
        day = x[8:10]
        
        if int(year) > 2011:
            return pd.to_datetime(f'{int(year)-100}-{month}-{day}', format='%Y-%m-%d')
        else:
            return pd.to_datetime(f'{year}-{month}-{day}', format='%Y-%m-%d')
    else:
        return x
```

```python
loan_data['earliest_cr_line_date'] = loan_data['earliest_cr_line_date'].apply(fix_date)
```

In fact we need the time since the earliest credit line was issued in order to use it in a regression. We have the date of earliest credit line issue.

Now we should set a reference date to calculate the time that has passed.For instance we can take today's date and that's what we often do in practice.

But since our data is older let's assume it's December 2017.

```python
#Create Months since earliest credit line
# Save in feature 
# make in month
loan_data['mths_since_earliest_cr_line'] = round((pd.to_datetime('2017-12-01') - loan_data['earliest_cr_line_date']) / 
                                                np.timedelta64(1, 'M'))
```

## Preprocessing discrete variables

We are going to preprocess the following discrete (categorical) variables: `grade`, `sub_grade`, `home_ownership`, `verification_status`, `loan_status`, `purpose`, `addr_state`, `initial_list_status`. Most likely, we are not going to use `sub_grade`, as it overlaps with grade.

- `grade` : LC assigned loan grade
- `sub_grade` : LC assigned loan subgrade
- `home_ownership` : The home ownership status provided by the borrower during registration. Our values are: RENT, OWN, MORTGAGE, OTHER
- `verification_status` : 
- `loan_status` : Current status of the loan
- `purpose` : A category provided by the borrower for the loan request. 
- `addr_state` : The state provided by the borrower in the loan application
- `initial_list_status` : The initial listing status of the loan. Possible values are â€“ W, F

```python
pd.get_dummies(loan_data['grade'], prefix='grade', prefix_sep=':')
```

```python
cols = ['grade', 'sub_grade', 'home_ownership', 'verification_status', 'loan_status', 'purpose', 
        'addr_state', 'initial_list_status']

dscrt_var_dummies = []

for col in cols:
    dscrt_var_dummies.append(pd.get_dummies(loan_data[col], prefix=col, prefix_sep=':'))
    
dscrt_var_dummies
```

```python
dscrt_var_dummies = pd.concat(dscrt_var_dummies, axis=1)
dscrt_var_dummies
```

```python
#Concat wtih `loan_data`

loan_data = pd.concat([loan_data, dscrt_var_dummies], axis=1)
```
